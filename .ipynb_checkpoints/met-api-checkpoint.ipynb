```python
import pandas as pd
import requests
import json

met = pd.read_csv('met.csv')

series = met[met.columns[1]].dropna()
series.head(5)
```

```python
api_url = []
for u in series:
    number = (str)(u.split('/')[-1])
    api_url.append('https://collectionapi.metmuseum.org/public/collection/v1/objects/' + number)

met['API_URL'] = pd.Series(api_url)
api_url_series = met['API_URL'].dropna()
print(api_url_series.head(500))
```

```python
image_url_json = []
for url in api_url_series:
    response = requests.get(url)
    image_url = {}
    if response.status_code == 200:
        additionalImages = []
        image_url["objectID"] = response.json()['objectID']
        image_url["title"] = response.json()['title']
        image_url["primaryImage"] = response.json()['primaryImage']
        additionalImages = response.json()['additionalImages']
        for (index, image) in enumerate(additionalImages):
            columnName = "Additional Images " + (str)(index)
            image_url[columnName] = image
        image_url_json.append(image_url)

print(image_url_json)

json_file = json.dumps(image_url_json, indent=2)
pd.read_json(json_file).to_csv("met_image_url_1.csv", index=False, encoding="utf-8")
```

```python
image_csv = pd.read_csv('met_image_url.csv')
print(image_csv.head(500))

for (index, row) in image_csv.iterrows():
    response = requests.get(row['primaryImage'])
    if response.status_code == 200:
        with open("/Users/qum/Documents/GitHub/chinese-art/Images/MET/" + (str)(row['objectID']) + ".jpg", 'wb') as f:
            f.write(response.content)
```
